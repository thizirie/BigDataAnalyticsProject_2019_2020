```python
import pandas as pd
import re 
import json
import ast
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, f1_score, classification_report, precision_score
import matplotlib.pylab as plt # plotting
```

```python
df = pd.read_csv('ted_main_dummy_columns.csv')
```

A function to spit a list into num groups, in such a way that each group has the same number of items.

```python
def chunkIt(seq, num):
    avg = len(seq) / float(num)
    out = []
    last = 0.0

    while last < len(seq):
        out.append(seq[int(last):int(last + avg)])
        last += avg

    return out
```

The next three boxes are for showing the range of each group.

```python
chunkIt(df.speaker_trend.sort_values().tolist(), 3)[0]
```

```python
chunkIt(df.speaker_trend.sort_values().tolist(), 3)[1]
```

```python
chunkIt(df.speaker_trend.sort_values().tolist(), 3)[2]
```

Creating a new column for our target class.

```python
df['speaker_class']=0
```

Setting the speaker_class according to the ranges we got from last three boxes.

```python
df.loc[((df.speaker_trend >= 0)&(df.speaker_trend <= 5.1111111110000005)), 'speaker_class'] = 0
df.loc[((df.speaker_trend > 5.1111111110000005)&(df.speaker_trend <  17.15277778)), 'speaker_class'] = 1 #removed = from numbers
df.loc[((df.speaker_trend >=17.15277778)&(df.speaker_trend <= 100)), 'speaker_class'] = 2
```

Seeing the confusion matrix between popularity_class and speaker_class.

```python
print(classification_report(df['popularity_class'], df['speaker_class']))
confusion_matrix(df['popularity_class'], df['speaker_class'])
```

Seeing precision between popularity_class and speaker_class.

```python
print("Precision score", precision_score(df['popularity_class'], df['speaker_class'],average='micro'))
print("Accuracy",accuracy_score(df['popularity_class'], df['speaker_class']))
```
