A function to spit a list into num groups, in such a way that each group has the same number of items.

```python
def chunkIt(seq, num):
    avg = len(seq) / float(num)
    out = []
    last = 0.0

    while last < len(seq):
        out.append(seq[int(last):int(last + avg)])
        last += avg

    return out
```

The next three boxes are for showing the range of each group.

```python
chunkIt(df.popularity.sort_values().tolist(), 3)[0]
```

```python
chunkIt(df.popularity.sort_values().tolist(), 3)[1]
```

```python
chunkIt(df.popularity.sort_values().tolist(), 3)[2]
```
